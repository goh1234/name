```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('weatherHistory.csv')
```

```python
data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
df = data[['Humidity', 'Wind Speed (km/h)', 'Pressure (millibars)', 'Temperature (C)', 'Daily Summary']]
```

```python
df = df.dropna()
```

```python
x = df[['Humidity', 'Wind Speed (km/h)', 'Pressure (millibars)']]
```

```python
x = sm.add_constant(x)
```

```python
y = data['Temperature (C)']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = sm.OLS(y, np.asarray(x)).fit()
print(model.summary())
```

```python
model = LinearRegression()
```

```python
model.fit(x.values, y)
```

```python
model.score(x.values, y)
```

```python
print('Humidity = 0.9, Wind Speed = 15, Pressure = 1000, 인 경우 Temperature =',
      round(model.predict([[0, 0.9, 15, 1000]])[0], 3), '이다.')
```
