```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris
```

```python
feature = iris['data']
```

```python
labels = iris['target']
```

```python
feature_names = iris['feature_names']
```

```python
df = pd.DataFrame(feature, columns=feature_names)
```

```python
df['target'] = labels
```

```python
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
```

```python
plt.figure(figsize = (6,4))
x = np.arange(len(df0))
plt.bar(x, df0['sepal width (cm)'])
plt.title('Sepal width bar graph of iris setosa', fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (6,4))
fig, ax = plt.subplots()

ax.boxplot(df0['sepal width (cm)'])
plt.show()
```

```python
plt.figure(figsize = (20,4))

ratio = [len(df0), len(df1), len(df2)]
labels = ['setosa', 'versicolor', 'virginica']

plt.pie(ratio, labels = labels)
plt.show()
```

```python
plt.figure(figsize = (20,4))

ratio = [len(df0), len(df1), len(df2)]
labels = ['setosa', 'versicolor', 'virginica']

plt.pie(ratio, labels = labels, autopct='%.1f%%')
plt.show()
```

```python
plt.figure(figsize = (20,4))

ratio = [len(df0), len(df1), len(df2)]
labels = ['setosa', 'versicolor', 'virginica']

plt.pie(ratio, labels = labels, autopct='%.1f%%')
plt.legend(loc='best', bbox_to_anchor=(1,1))
plt.show()
```

```python
plt.figure(figsize = (6,4))
fig, ax = plt.subplots()

ax.boxplot([df0['sepal length (cm)'], df1['sepal length (cm)'], df2['sepal length (cm)']])
plt.title('Sepal length boxplots of three types of iris', fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (6,4))

plt.hist(df0['sepal length (cm)'], alpha = 0.5, label = 'setosa')
plt.hist(df1['sepal length (cm)'], alpha = 0.5, label = 'versicolor')
plt.hist(df2['sepal length (cm)'], alpha = 0.5, label = 'virginca')

plt.title('Sepal length histograms of three types of iris', fontsize = 14)

plt.legend()

plt.show()
```
