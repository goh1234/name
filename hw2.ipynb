```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
feature = iris['data']
```

```python
labels = iris['target']
```

```python
feature_names = iris['feature_names']
```

```python
df = pd.DataFrame(feature, columns=feature_names)
```

```python
df['target'] = labels
```

```python
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
```

```python
plt.figure(figsize = (6,4))
x = np.arange(len(df0))
plt.bar(x, df0['petal length (cm)'])
plt.title('Petal length bar graph of iris setosa', fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (6,4))

plt.hist(df0['petal length (cm)'], alpha = 0.5, label = 'setosa')
plt.hist(df1['petal length (cm)'], alpha = 0.5, label = 'versicolor')
plt.hist(df2['petal length (cm)'], alpha = 0.5, label = 'virginca')

plt.title('Petal length histograms of three types of iris', fontsize = 14)

plt.legend()

plt.show()
```

```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
wine
```

```python
feature = wine['data']
```

```python
labels = wine['target']
```

```python
feature_names = wine['feature_names']
```

```python
df = pd.DataFrame(feature, columns=feature_names)
```

```python
df['target'] = labels
```

```python
df
```

```python
df0 = df[df['target'] == 0]
df1 = df[df['target'] == 1]
df2 = df[df['target'] == 2]
```

```python
plt.figure(figsize = (20,4))

ratio = [len(df0), len(df1), len(df2)]
labels = ['class_0', 'class_1', 'class_2']

plt.pie(ratio, labels = labels, autopct='%.1f%%')
plt.legend(loc='best', bbox_to_anchor=(1.1,1))
plt.show()
```

```python
plt.figure(figsize = (6,4))
fig, ax = plt.subplots()
labels = ['class_0', 'class_1', 'class_2']
ax.boxplot([df0['alcohol'], df1['alcohol'], df2['alcohol']], tick_labels = labels)
plt.title('Alcohol levels of three types of wine', fontsize = 14)
plt.show()
```

```python
plt.figure(figsize = (6,4))

plt.hist(df0['alcohol'], alpha = 0.5, label = 'class_0')
plt.hist(df1['alcohol'], alpha = 0.5, label = 'class_1')
plt.hist(df2['alcohol'], alpha = 0.5, label = 'virginca')

plt.title('Alcohol level histograms of three types of wine', fontsize = 14)

plt.legend()

plt.show()
```
