```python
import pandas as pd
import statistics
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
#1. 필요 없는 column을 지워보세요.
```

```python
df.drop(['PassengerId', 'Survived', 'Pclass', 'Ticket', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin'], axis = 1, inplace=True)
```

```python
#2. 이번엔 age는 mean으로, Embarked는 median으로 imputation해보세요.
```

```python
df['Age'].fillna(df['Age'].mean(), inplace = True)
```

```python
df['Embarked'].fillna(df['Embarked'].median(), inplace = True)
```

```python
df
```

```python
#3. Sex, Embarked column을 정수 인코딩 해보세요.
```

```python
embarked_mapping = {'Q':0, 'S':1, 'C':2}
df['Embarked'] = df['Embarked'].map(embarked_mapping)
```

```python
Sex_mapping = {'male':0, 'female':1}
df['Sex'] = df['Sex'].map(Sex_mapping)
```
