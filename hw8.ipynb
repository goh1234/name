```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df['Age'] = df['Age'].fillna((df['Age'].mean()))
```

```python
df
```

```python
X = df.iloc[:, [2,5,6,7,9]]
X
```

```python
y = df.iloc[:, 1]
y
```

```python
X = X.values
y = y.values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=2026)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train  = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
rf_ne_10 = RandomForestClassifier(n_estimators=10)
rf_ne_100 = RandomForestClassifier(n_estimators=100)
```

```python
rf_ne_10.fit(X_train, y_train)
rf_ne_100.fit(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
knn_3 = KNeighborsClassifier(n_neighbors=3)
knn_4 = KNeighborsClassifier(n_neighbors=4)
knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_6 = KNeighborsClassifier(n_neighbors=6)
knn_7 = KNeighborsClassifier(n_neighbors=7)
```

```python
knn_3.fit(X_train, y_train)
knn_4.fit(X_train, y_train)
knn_5.fit(X_train, y_train)
knn_6.fit(X_train, y_train)
knn_7.fit(X_train, y_train)
```

```python
print(f'accuracy of KNN (k=3) is', round(np.mean(knn_3.predict(X_test)==y_test), 2))
print(f'accuracy of KNN (k=4) is', round(np.mean(knn_4.predict(X_test)==y_test), 2))
print(f'accuracy of KNN (k=5) is', round(np.mean(knn_5.predict(X_test)==y_test), 2))
print(f'accuracy of KNN (k=6) is', round(np.mean(knn_6.predict(X_test)==y_test), 2))
print(f'accuracy of KNN (k=7) is', round(np.mean(knn_7.predict(X_test)==y_test), 2))
```
