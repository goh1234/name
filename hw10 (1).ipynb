```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1.0, -2.0])
w = np.array([[0.5, -1.0, 2.0],
              [1.5,  0.0, -0.5]])
b = np.array([0.1, -0.2, 0.3])
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def step(x):
    return (x>0).astype(int)
```

```python
x = x.reshape(-1,1)
```

```python
x.shape
```

```python
sigmoid(x*w + b)
```

```python
relu(x*w + b)
```

```python
step(x*w + b)
```

```python
y_observed = [0.06, 0.8, 0.11, 0.65, 0.34, 0.5, 0.74, 0.62, 0.4]
y_predicted = [0.05, 0.9, 0.1, 0.6, 0.2, 0.8, 0.3, 0.7, 0.5]
```

```python
def mae(y_observed, y_predicted):
    y_observed_array = np.array(y_observed)
    y_predicted_array = np.array(y_predicted)
    return np.mean((y_observed_array - y_predicted_array))
```

```python
mae(y_observed, y_predicted)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_observed, y_predicted)
```
