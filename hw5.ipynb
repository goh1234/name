```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
data
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
x = data.YearBuilt
```

```python
y = data.LotArea
```

```python
x.values.shape
y.values.shape
```

```python
x = sm.add_constant(x)
```

```python
model = LinearRegression()
```

```python
model.fit(x.values, y)
```

```python
model.predict(np.array([[1, 2026]]))[0]
```

```python
year = 2026
print(f'the area of house built in {year} is {model.predict(np.array([[1,year]]))[0]}')
```
